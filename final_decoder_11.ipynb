```python
import os
import json
import random
import logging
from PIL import Image

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import timm
import cv2
from sklearn.model_selection import train_test_split
from transformers import GPT2Tokenizer, GPT2LMHeadModel, GPT2Config
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

# for semantic similarity
from sentence_transformers import SentenceTransformer, util

# ----------------------------
# 1. Dataset & Collate
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer, image_transform=None, patch_transform=None):
        self.tokenizer = tokenizer
        # Default image transform with mild geometric augmentations
        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(
                degrees=10,             # Â±10Â°
                translate=(0.1, 0.1),   # Â±10% translation
                scale=(0.95, 1.05)      # 0.95Ã—â€“1.05Ã— scaling
            ),
            transforms.ToTensor()
        ])
        # Default patch transform with the same mild augmentations
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(
                degrees=10,
                translate=(0.1, 0.1),
                scale=(0.95, 1.05)
            ),
            transforms.ToTensor()
        ])
        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path','')
            fps = item.get('file_paths',[])
            if isinstance(fps,str): fps=[fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged,fps,item))

        self.data = {}
        for i,(merged,fps,item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints',{}),
                'clean': item.get('cleaned_report','')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        # full image
        img = Image.open(e['full']).convert('RGB')
        img = self.image_transform(img)
        # patches
        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [ self.patch_transform(Image.fromarray(p)) for p in patches ]
        patches_tensor = torch.stack(pt, 0) if pt else torch.zeros(34, 3, 112, 112)
        # tokenize
        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(prompt, truncation=True, max_length=512, return_tensors='pt')
        input_ids      = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict, crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst=[]; pts=side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2,0)
                    y0 = max(y-crop_size[1]//2,0)
                    x1 = min(x+crop_size[0]//2,arr.shape[1])
                    y1 = min(y+crop_size[1]//2,arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size: lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1],patch_size[0],3),np.uint8)
            while len(lst)<17: lst.append(black)
            return lst[:17]

        left,right = [],[]
        if len(paths)==1:
            p=paths[0]
            if os.path.exists(p):
                arr=cv2.cvtColor(cv2.imread(p),cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):  left  = extract(arr,kps_dict['left'])
                if kps_dict.get('right'): right = extract(arr,kps_dict['right'])
        else:
            for side,p in zip(['left','right'],paths):
                if p and os.path.exists(p):
                    arr=cv2.cvtColor(cv2.imread(p),cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst=extract(arr,kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right: right=[cv2.flip(p,1) for p in left]
        if right and not left: left=[cv2.flip(p,1) for p in right]
        if not left and not right: return pad17([])+pad17([])
        return pad17(left)+pad17(right)

def collate_fn(batch):
    full_imgs = torch.stack([b['full_img'] for b in batch])
    patches   = torch.stack([b['patches']   for b in batch])
    ids       = [b['input_ids']      for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    clean_txt = [b['clean_text']     for b in batch]

    input_ids      = nn.utils.rnn.pad_sequence(ids, batch_first=True, padding_value=0)
    attention_mask = nn.utils.rnn.pad_sequence(masks, batch_first=True, padding_value=0)

    return {
        'full_img':       full_imgs,
        'patches':        patches,
        'input_ids':      input_ids,
        'attention_mask': attention_mask,
        'clean_text':     clean_txt
    }

# ----------------------------
# 2. Model
# ----------------------------
class VisionFusionEncoder(nn.Module):
    def __init__(self, vision_dim=768, num_heads=8):
        super().__init__()
        self.global_enc   = timm.create_model('swin_base_patch4_window7_224', pretrained=True, num_classes=0)
        self.global_proj  = nn.Linear(self.global_enc.num_features, vision_dim)
        res = models.resnet50(pretrained=True); res.fc = nn.Identity()
        self.patch_enc    = res
        self.patch_proj   = nn.Linear(2048, vision_dim)
        # two cross-attention layers for deeper fusion
        self.cross_attn1  = nn.MultiheadAttention(vision_dim, num_heads, batch_first=True)
        self.norm1        = nn.LayerNorm(vision_dim)
        self.cross_attn2  = nn.MultiheadAttention(vision_dim, num_heads, batch_first=True)
        self.norm2        = nn.LayerNorm(vision_dim)

    def forward(self, full_img, patches):
        B = full_img.size(0)
        g = self.global_proj(self.global_enc(full_img))           # BÃ—D
        P = patches.view(-1,3,112,112)
        p = self.patch_proj(self.patch_enc(P)).view(B,34,-1)      # BÃ—34Ã—D
        # first fusion
        q1, _ = self.cross_attn1(g.unsqueeze(1), p, p)            # BÃ—1Ã—D
        fused1 = self.norm1(q1.squeeze(1) + g)                     # BÃ—D
        # second fusion
        q2, _ = self.cross_attn2(fused1.unsqueeze(1), p, p)       # BÃ—1Ã—D
        fused2 = self.norm2(q2.squeeze(1) + fused1)                # BÃ—D
        return torch.cat([fused2.unsqueeze(1), p], dim=1)         # BÃ—35Ã—D

class MedicalReportGenerator(nn.Module):
    def __init__(self, tokenizer, vision_dim=768, gpt2_model='gpt2', device='cuda', sem_coef=0.1):
        super().__init__()
        self.device    = device
        self.tokenizer = tokenizer
        self.embedder  = SentenceTransformer('all-MiniLM-L6-v2')
        self.sem_coef  = sem_coef
        self.encoder   = VisionFusionEncoder(vision_dim)
        cfg = GPT2Config.from_pretrained(gpt2_model, add_cross_attention=True)
        self.decoder   = GPT2LMHeadModel.from_pretrained(gpt2_model, config=cfg)
        self.to(device)

    def forward(self, full_img, patches, input_ids, attention_mask, clean_texts, labels=None):
        # encode vision
        vt = self.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=self.device)
        # decode with cross-attention
        outputs = self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            labels=labels,
            return_dict=True
        )
        ce_loss = outputs.loss

        # semantic-aware loss
        with torch.no_grad():
            gen_ids = outputs.logits.argmax(-1)
            gen_texts = [ self.tokenizer.decode(g, skip_special_tokens=True) for g in gen_ids ]
            emb_gen   = self.embedder.encode(
                gen_texts,
                convert_to_tensor=True,
                device=self.device,
                show_progress_bar=False
            )
            emb_truth = self.embedder.encode(
                clean_texts,
                convert_to_tensor=True,
                device=self.device,
                show_progress_bar=False
            )
        sims      = util.cos_sim(emb_gen, emb_truth)  # (batch, batch)
        sem_loss  = torch.mean(1.0 - sims.diag())

        loss = ce_loss + self.sem_coef * sem_loss
        return loss, outputs.logits

# ----------------------------
# 3. Training & Eval
# ----------------------------
def train_epoch(model, loader, optim, scheduler=None, epoch=0):
    model.train()
    tot_loss, tot_acc, n = 0.0, 0.0, 0
    loop = tqdm(loader, desc=f"Epoch {epoch} [Train]", leave=False)
    for batch in loop:
        full    = batch['full_img'].to(model.device)
        patches = batch['patches'].to(model.device)
        ids     = batch['input_ids'].to(model.device)
        mask    = batch['attention_mask'].to(model.device)
        clean   = batch['clean_text']

        optim.zero_grad()
        loss, logits = model(full, patches, ids, mask, clean, labels=ids)
        loss.backward()
        optim.step()
        if scheduler: scheduler.step()

        preds = logits.argmax(-1)
        acc = (preds == ids).masked_select(mask.bool()).float().mean().item()

        bsz = ids.size(0)
        tot_loss += loss.item() * bsz
        tot_acc  += acc * bsz
        n       += bsz
        loop.set_postfix(loss=loss.item(), acc=acc)

    return tot_loss / n, tot_acc / n

@torch.no_grad()
def eval_epoch(model, loader, epoch=0):
    model.eval()
    tot_loss, tot_acc, n = 0.0, 0.0, 0
    loop = tqdm(loader, desc=f"Epoch {epoch} [Val]  ", leave=False)
    for batch in loop:
        full    = batch['full_img'].to(model.device)
        patches = batch['patches'].to(model.device)
        ids     = batch['input_ids'].to(model.device)
        mask    = batch['attention_mask'].to(model.device)
        clean   = batch['clean_text']

        loss, logits = model(full, patches, ids, mask, clean, labels=ids)
        preds = logits.argmax(-1)
        acc = (preds == ids).masked_select(mask.bool()).float().mean().item()

        bsz = ids.size(0)
        tot_loss += loss.item() * bsz
        tot_acc  += acc * bsz
        n       += bsz
        loop.set_postfix(loss=loss.item(), acc=acc)

    return tot_loss / n, tot_acc / n

# ----------------------------
# 4. Plotting
# ----------------------------
def plot_metrics(train_losses, val_losses, train_accs, val_accs):
    epochs = range(1, len(train_losses) + 1)
    plt.figure(); plt.plot(epochs, train_losses);   plt.plot(epochs, val_losses)
    plt.title('Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(['Train','Val']); plt.show()
    plt.figure(); plt.plot(epochs, train_accs);   plt.plot(epochs, val_accs)
    plt.title('Token-Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(['Train','Val']); plt.show()

# ----------------------------
# 5. Main
# ----------------------------
def main():
    logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s | %(levelname)s | %(message)s',
    handlers=[
        logging.FileHandler("training.log"),
        logging.StreamHandler()
    ]
)
    logging.basicConfig(level=logging.INFO)
    JSON_PATH = './final_samples_both_only_v2_cleaned.json'
    DEVICE    = 'cuda' if torch.cuda.is_available() else 'cpu'
    BATCH, EPOCHS, LR = 8, 1, 1e-4

    # tokenizer
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token': '<|pad|>'})
    tokenizer.bos_token = tokenizer.eos_token

    # data & splits
    ds = FinalSamplesDataset(JSON_PATH, tokenizer)
    idxs = list(range(len(ds)))
    tr, tmp = train_test_split(idxs, test_size=0.2, random_state=42)
    va, te  = train_test_split(tmp, test_size=0.5, random_state=42)

    train_loader = DataLoader(torch.utils.data.Subset(ds, tr),
                              batch_size=BATCH, shuffle=True, collate_fn=collate_fn)
    val_loader   = DataLoader(torch.utils.data.Subset(ds, va),
                              batch_size=BATCH, shuffle=False, collate_fn=collate_fn)
    test_loader  = DataLoader(torch.utils.data.Subset(ds, te),
                              batch_size=1, shuffle=False, collate_fn=collate_fn)

    # model & optimizer
    model = MedicalReportGenerator(tokenizer, device=DEVICE)
    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)

    # train & validate
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    for ep in range(1, EPOCHS + 1):
        tr_l, tr_a = train_epoch(model, train_loader, optimizer, epoch=ep)
        va_l, va_a = eval_epoch(model, val_loader, epoch=ep)
        print(f"Epoch {ep:02d} â†’ train_loss={tr_l:.4f}, val_loss={va_l:.4f}, train_acc={tr_a:.4f}, val_acc={va_a:.4f}")
        train_losses.append(tr_l)
        val_losses.append(va_l)
        train_accs.append(tr_a)
        val_accs.append(va_a)

    plot_metrics(train_losses, val_losses, train_accs, val_accs)

    # ----------------------------
    # 6. Test: sampling + overall similarity
    # ----------------------------
    model.eval()

    print("\n20 random examples:")
    examples = random.sample(range(len(te)), 20)
    for idx in examples:
        batch = next(iter(DataLoader([ds[te[idx]]], batch_size=1, collate_fn=collate_fn)))
        full_img  = batch['full_img'].to(DEVICE)
        patches   = batch['patches'].to(DEVICE)
        clean_txt = batch['clean_text'][0]

        vt = model.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=DEVICE)

        prefix = tokenizer(f"{tokenizer.bos_token} FINDINGS:", return_tensors='pt').to(DEVICE)
        out = model.decoder.generate(
            **prefix,
            max_length=128,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=1.0,
            num_return_sequences=1,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            use_cache=True
        )
        gen = tokenizer.decode(out[0], skip_special_tokens=True)

        emb = model.embedder.encode([gen, clean_txt], convert_to_tensor=True, device=DEVICE)
        sim = util.cos_sim(emb[0], emb[1]).item()

        print("â€”" * 40)
        print("Clean     :", clean_txt)
        print("Generated :", gen)
        print(f"Similarity: {sim:.4f}")

    # overall average
    sims = []
    for batch in tqdm(test_loader, desc="Evaluating test set"):
        full_img  = batch['full_img'].to(DEVICE)
        patches   = batch['patches'].to(DEVICE)
        clean_txt = batch['clean_text'][0]

        vt = model.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=DEVICE)

        prefix = tokenizer(f"{tokenizer.bos_token} FINDINGS:", return_tensors='pt').to(DEVICE)
        out = model.decoder.generate(
            **prefix,
            max_length=128,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=1.0,
            num_return_sequences=1,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            use_cache=True
        )
        gen = tokenizer.decode(out[0], skip_special_tokens=True)
        emb_pair = model.embedder.encode([gen, clean_txt], convert_to_tensor=True, device=DEVICE)
        sims.append(util.cos_sim(emb_pair[0], emb_pair[1]).item())

    overall = float(np.mean(sims))
    logging.info(f"\nAverage semantic similarity over {len(sims)} test samples: {overall:.4f}")

if __name__ == '__main__':
    main()
```

```python
import os
import json
import random
import logging
from PIL import Image

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import timm
import cv2
from sklearn.model_selection import train_test_split
from transformers import GPT2Tokenizer, GPT2LMHeadModel, GPT2Config
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

# for semantic similarity
from sentence_transformers import SentenceTransformer, util

# ----------------------------
# 1. Dataset & Collate
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer, image_transform=None, patch_transform=None):
        self.tokenizer = tokenizer
        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(
                degrees=10, translate=(0.1, 0.1), scale=(0.95, 1.05)
            ),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(
                degrees=10, translate=(0.1, 0.1), scale=(0.95, 1.05)
            ),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        # prefer merged if it exists, otherwise first FPS
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt, 0) if pt else torch.zeros(34, 3, 112, 112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(prompt, truncation=True, max_length=512, return_tensors='pt')
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict, crop_size=(200, 300), patch_size=(112, 112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x, y, s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s > 0:
                    x0 = max(x - crop_size[0]//2, 0)
                    y0 = max(y - crop_size[1]//2, 0)
                    x1 = min(x + crop_size[0]//2, arr.shape[1])
                    y1 = min(y + crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths) == 1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left', 'right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        if side == 'left':
                            left.extend(lst)
                        else:
                            right.extend(lst)

        if left and not right:
            right = [cv2.flip(p, 1) for p in left]
        if right and not left:
            left = [cv2.flip(p, 1) for p in right]
        if not left and not right:
            return pad17([]) + pad17([])
        return pad17(left) + pad17(right)

def collate_fn(batch):
    full_imgs = torch.stack([b['full_img'] for b in batch])
    patches   = torch.stack([b['patches']   for b in batch])
    ids       = [b['input_ids']      for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    clean_txt = [b['clean_text']     for b in batch]

    input_ids      = nn.utils.rnn.pad_sequence(ids, batch_first=True, padding_value=0)
    attention_mask = nn.utils.rnn.pad_sequence(masks, batch_first=True, padding_value=0)

    return {
        'full_img':       full_imgs,
        'patches':        patches,
        'input_ids':      input_ids,
        'attention_mask': attention_mask,
        'clean_text':     clean_txt
    }


# ----------------------------
# 2. Model
# ----------------------------
class VisionFusionEncoder(nn.Module):
    def __init__(self, vision_dim=768, num_heads=8):
        super().__init__()
        self.global_enc   = timm.create_model('swin_base_patch4_window7_224', pretrained=True, num_classes=0)
        self.global_proj  = nn.Linear(self.global_enc.num_features, vision_dim)
        res = models.resnet50(pretrained=True)
        res.fc = nn.Identity()
        self.patch_enc    = res
        self.patch_proj   = nn.Linear(2048, vision_dim)
        self.cross_attn1  = nn.MultiheadAttention(vision_dim, num_heads, batch_first=True)
        self.norm1        = nn.LayerNorm(vision_dim)
        self.cross_attn2  = nn.MultiheadAttention(vision_dim, num_heads, batch_first=True)
        self.norm2        = nn.LayerNorm(vision_dim)

    def forward(self, full_img, patches):
        B = full_img.size(0)
        g = self.global_proj(self.global_enc(full_img))           # BÃ—D
        P = patches.view(-1,3,112,112)
        p = self.patch_proj(self.patch_enc(P)).view(B,34,-1)      # BÃ—34Ã—D

        q1, _ = self.cross_attn1(g.unsqueeze(1), p, p)            # BÃ—1Ã—D
        fused1 = self.norm1(q1.squeeze(1) + g)                     # BÃ—D

        q2, _ = self.cross_attn2(fused1.unsqueeze(1), p, p)       # BÃ—1Ã—D
        fused2 = self.norm2(q2.squeeze(1) + fused1)               # BÃ—D

        return torch.cat([fused2.unsqueeze(1), p], dim=1)         # BÃ—35Ã—D


class MedicalReportGenerator(nn.Module):
    def __init__(self, tokenizer, vision_dim=768, gpt2_model='gpt2', device='cuda', sem_coef=0.1):
        super().__init__()
        self.device    = device
        self.tokenizer = tokenizer
        self.embedder  = SentenceTransformer('all-MiniLM-L6-v2').to(device)
        self.sem_coef  = sem_coef
        self.encoder   = VisionFusionEncoder(vision_dim)

        cfg = GPT2Config.from_pretrained(gpt2_model, add_cross_attention=True)
        self.decoder   = GPT2LMHeadModel.from_pretrained(gpt2_model, config=cfg)

        # **CRITICAL**: resize embeddings to account for the new pad token
        self.decoder.resize_token_embeddings(len(self.tokenizer))

        self.to(device)

    def forward(self, full_img, patches, input_ids, attention_mask, clean_texts, labels=None):
        vt = self.encoder(full_img, patches)                         # BÃ—35Ã—D
        vm = torch.ones(vt.size()[:2], device=self.device)

        outputs = self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            labels=labels,
            return_dict=True
        )
        ce_loss = outputs.loss

        # compute semantic loss with gradient
        gen_ids = outputs.logits.argmax(-1)
        gen_texts = [self.tokenizer.decode(g, skip_special_tokens=True) for g in gen_ids]

        emb_gen   = self.embedder.encode(
            gen_texts,
            convert_to_tensor=True,
            device=self.device,
            show_progress_bar=False
        )
        emb_truth = self.embedder.encode(
            clean_texts,
            convert_to_tensor=True,
            device=self.device,
            show_progress_bar=False
        )

        sims      = util.cos_sim(emb_gen, emb_truth)  # (batch, batch)
        sem_loss  = torch.mean(1.0 - sims.diag())

        loss = ce_loss + self.sem_coef * sem_loss
        return loss, outputs.logits


# ----------------------------
# 3. Training & Eval
# ----------------------------
def train_epoch(model, loader, optim, scheduler=None, epoch=0):
    model.train()
    tot_loss, tot_acc, n = 0.0, 0.0, 0
    loop = tqdm(loader, desc=f"Epoch {epoch} [Train]", leave=False)
    for batch in loop:
        full    = batch['full_img'].to(model.device)
        patches = batch['patches'].to(model.device)
        ids     = batch['input_ids'].to(model.device)
        mask    = batch['attention_mask'].to(model.device)
        clean   = batch['clean_text']

        optim.zero_grad()
        loss, logits = model(full, patches, ids, mask, clean, labels=ids)
        loss.backward()
        optim.step()
        if scheduler: scheduler.step()

        preds = logits.argmax(-1)
        acc = (preds == ids).masked_select(mask.bool()).float().mean().item()

        bsz = ids.size(0)
        tot_loss += loss.item() * bsz
        tot_acc  += acc * bsz
        n       += bsz
        loop.set_postfix(loss=loss.item(), acc=acc)

    return tot_loss / n, tot_acc / n

@torch.no_grad()
def eval_epoch(model, loader, epoch=0):
    model.eval()
    tot_loss, tot_acc, n = 0.0, 0.0, 0
    loop = tqdm(loader, desc=f"Epoch {epoch} [Val]  ", leave=False)
    for batch in loop:
        full    = batch['full_img'].to(model.device)
        patches = batch['patches'].to(model.device)
        ids     = batch['input_ids'].to(model.device)
        mask    = batch['attention_mask'].to(model.device)
        clean   = batch['clean_text']

        loss, logits = model(full, patches, ids, mask, clean, labels=ids)
        preds = logits.argmax(-1)
        acc = (preds == ids).masked_select(mask.bool()).float().mean().item()

        bsz = ids.size(0)
        tot_loss += loss.item() * bsz
        tot_acc  += acc * bsz
        n       += bsz
        loop.set_postfix(loss=loss.item(), acc=acc)

    return tot_loss / n, tot_acc / n


# ----------------------------
# 4. Plotting
# ----------------------------
def plot_metrics(train_losses, val_losses, train_accs, val_accs):
    epochs = range(1, len(train_losses) + 1)
    plt.figure()
    plt.plot(epochs, train_losses)
    plt.plot(epochs, val_losses)
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train','Val'])
    plt.show()

    plt.figure()
    plt.plot(epochs, train_accs)
    plt.plot(epochs, val_accs)
    plt.title('Token-Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train','Val'])
    plt.show()


# ----------------------------
# 5. Main
# ----------------------------
def main():
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s | %(levelname)s | %(message)s',
        handlers=[
            logging.FileHandler("training.log"),
            logging.StreamHandler()
        ]
    )

    JSON_PATH = './final_samples_both_only_v2_cleaned.json'
    DEVICE    = 'cuda' if torch.cuda.is_available() else 'cpu'
    BATCH, EPOCHS, LR = 8, 1, 1e-4

    # tokenizer
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token': '<|pad|>'})
    tokenizer.bos_token = tokenizer.eos_token

    # data & splits
    ds = FinalSamplesDataset(JSON_PATH, tokenizer)
    idxs = list(range(len(ds)))
    tr, tmp = train_test_split(idxs, test_size=0.2, random_state=42)
    va, te  = train_test_split(tmp, test_size=0.5, random_state=42)

    train_loader = DataLoader(torch.utils.data.Subset(ds, tr),
                              batch_size=BATCH, shuffle=True, collate_fn=collate_fn)
    val_loader   = DataLoader(torch.utils.data.Subset(ds, va),
                              batch_size=BATCH, shuffle=False, collate_fn=collate_fn)
    test_loader  = DataLoader(torch.utils.data.Subset(ds, te),
                              batch_size=1, shuffle=False, collate_fn=collate_fn)

    # model & optimizer
    model     = MedicalReportGenerator(tokenizer, device=DEVICE)
    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)

    # train & validate
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    for ep in range(1, EPOCHS + 1):
        tr_l, tr_a = train_epoch(model, train_loader, optimizer, epoch=ep)
        va_l, va_a = eval_epoch(model, val_loader, epoch=ep)
        print(f"Epoch {ep:02d} â†’ train_loss={tr_l:.4f}, val_loss={va_l:.4f}, train_acc={tr_a:.4f}, val_acc={va_a:.4f}")
        train_losses.append(tr_l)
        val_losses.append(va_l)
        train_accs.append(tr_a)
        val_accs.append(va_a)

    plot_metrics(train_losses, val_losses, train_accs, val_accs)

    # ----------------------------
    # 6. Test: sampling + overall similarity
    # ----------------------------
    model.eval()

    print("\n20 random examples:")
    examples = random.sample(range(len(te)), 20)
    for idx in examples:
        batch = next(iter(DataLoader([ds[te[idx]]], batch_size=1, collate_fn=collate_fn)))
        full_img  = batch['full_img'].to(DEVICE)
        patches   = batch['patches'].to(DEVICE)
        clean_txt = batch['clean_text'][0]

        vt = model.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=DEVICE)

        prefix = tokenizer(f"{tokenizer.bos_token} FINDINGS:", return_tensors='pt').to(DEVICE)
        out = model.decoder.generate(
            **prefix,
            max_length=128,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=1.0,
            num_return_sequences=1,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            use_cache=True
        )
        gen = tokenizer.decode(out[0], skip_special_tokens=True)

        emb = model.embedder.encode([gen, clean_txt], convert_to_tensor=True, device=DEVICE)
        sim = util.cos_sim(emb[0], emb[1]).item()

        print("â€”" * 40)
        print("Clean     :", clean_txt)
        print("Generated :", gen)
        print(f"Similarity: {sim:.4f}")

    # overall average
    sims = []
    for batch in tqdm(test_loader, desc="Evaluating test set"):
        full_img  = batch['full_img'].to(DEVICE)
        patches   = batch['patches'].to(DEVICE)
        clean_txt = batch['clean_text'][0]

        vt = model.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=DEVICE)

        prefix = tokenizer(f"{tokenizer.bos_token} FINDINGS:", return_tensors='pt').to(DEVICE)
        out = model.decoder.generate(
            **prefix,
            max_length=128,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=1.0,
            num_return_sequences=1,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            use_cache=True
        )
        gen = tokenizer.decode(out[0], skip_special_tokens=True)
        emb_pair = model.embedder.encode([gen, clean_txt], convert_to_tensor=True, device=DEVICE)
        sims.append(util.cos_sim(emb_pair[0], emb_pair[1]).item())

    overall = float(np.mean(sims))
    logging.info(f"\nAverage semantic similarity over {len(sims)} test samples: {overall:.4f}")

if __name__ == '__main__':
    main()
```

```python
import os
import json
import random
import logging
from PIL import Image

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import timm
import cv2
from sklearn.model_selection import train_test_split
from transformers import GPT2Tokenizer, GPT2LMHeadModel, GPT2Config
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

# for semantic similarity
from sentence_transformers import SentenceTransformer, util

# ----------------------------
# 1. Dataset & Collate
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer, image_transform=None, patch_transform=None):
        self.tokenizer = tokenizer
        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(
                degrees=10, translate=(0.1, 0.1), scale=(0.95, 1.05)
            ),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(
                degrees=10, translate=(0.1, 0.1), scale=(0.95, 1.05)
            ),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        # prefer merged if it exists, otherwise first FPS
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt, 0) if pt else torch.zeros(34, 3, 112, 112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(prompt, truncation=True, max_length=512, return_tensors='pt')
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict, crop_size=(200, 300), patch_size=(112, 112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x, y, s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s > 0:
                    x0 = max(x - crop_size[0]//2, 0)
                    y0 = max(y - crop_size[1]//2, 0)
                    x1 = min(x + crop_size[0]//2, arr.shape[1])
                    y1 = min(y + crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths) == 1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left', 'right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        if side == 'left':
                            left.extend(lst)
                        else:
                            right.extend(lst)

        if left and not right:
            right = [cv2.flip(p, 1) for p in left]
        if right and not left:
            left = [cv2.flip(p, 1) for p in right]
        if not left and not right:
            return pad17([]) + pad17([])
        return pad17(left) + pad17(right)

def collate_fn(batch):
    full_imgs = torch.stack([b['full_img'] for b in batch])
    patches   = torch.stack([b['patches']   for b in batch])
    ids       = [b['input_ids']      for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    clean_txt = [b['clean_text']     for b in batch]

    input_ids      = nn.utils.rnn.pad_sequence(ids, batch_first=True, padding_value=0)
    attention_mask = nn.utils.rnn.pad_sequence(masks, batch_first=True, padding_value=0)

    return {
        'full_img':       full_imgs,
        'patches':        patches,
        'input_ids':      input_ids,
        'attention_mask': attention_mask,
        'clean_text':     clean_txt
    }


# ----------------------------
# 2. Model
# ----------------------------
class VisionFusionEncoder(nn.Module):
    def __init__(self, vision_dim=768, num_heads=8):
        super().__init__()
        self.global_enc   = timm.create_model('swin_base_patch4_window7_224', pretrained=True, num_classes=0)
        self.global_proj  = nn.Linear(self.global_enc.num_features, vision_dim)
        res = models.resnet50(pretrained=True)
        res.fc = nn.Identity()
        self.patch_enc    = res
        self.patch_proj   = nn.Linear(2048, vision_dim)
        self.cross_attn1  = nn.MultiheadAttention(vision_dim, num_heads, batch_first=True)
        self.norm1        = nn.LayerNorm(vision_dim)
        self.cross_attn2  = nn.MultiheadAttention(vision_dim, num_heads, batch_first=True)
        self.norm2        = nn.LayerNorm(vision_dim)

    def forward(self, full_img, patches):
        B = full_img.size(0)
        g = self.global_proj(self.global_enc(full_img))           # BÃ—D
        P = patches.view(-1,3,112,112)
        p = self.patch_proj(self.patch_enc(P)).view(B,34,-1)      # BÃ—34Ã—D

        q1, _ = self.cross_attn1(g.unsqueeze(1), p, p)            # BÃ—1Ã—D
        fused1 = self.norm1(q1.squeeze(1) + g)                     # BÃ—D

        q2, _ = self.cross_attn2(fused1.unsqueeze(1), p, p)       # BÃ—1Ã—D
        fused2 = self.norm2(q2.squeeze(1) + fused1)               # BÃ—D

        return torch.cat([fused2.unsqueeze(1), p], dim=1)         # BÃ—35Ã—D


class MedicalReportGenerator(nn.Module):
    def __init__(self, tokenizer, vision_dim=768, gpt2_model='gpt2', device='cuda', sem_coef=0.1):
        super().__init__()
        self.device    = device
        self.tokenizer = tokenizer
        self.embedder  = SentenceTransformer('all-MiniLM-L6-v2').to(device)
        self.sem_coef  = sem_coef
        self.encoder   = VisionFusionEncoder(vision_dim)

        cfg = GPT2Config.from_pretrained(gpt2_model, add_cross_attention=True)
        self.decoder   = GPT2LMHeadModel.from_pretrained(gpt2_model, config=cfg)
        # resize embeddings for the new pad token
        self.decoder.resize_token_embeddings(len(self.tokenizer))

        self.to(device)

    def forward(self, full_img, patches, input_ids, attention_mask, clean_texts, labels=None):
        vt = self.encoder(full_img, patches)                         # BÃ—35Ã—D
        vm = torch.ones(vt.size()[:2], device=self.device)

        outputs = self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            labels=labels,
            return_dict=True
        )
        ce_loss = outputs.loss

        # compute semantic loss with gradient
        gen_ids = outputs.logits.argmax(-1)
        gen_texts = [self.tokenizer.decode(g, skip_special_tokens=True) for g in gen_ids]

        emb_gen   = self.embedder.encode(
            gen_texts,
            convert_to_tensor=True,
            device=self.device,
            show_progress_bar=False
        )
        emb_truth = self.embedder.encode(
            clean_texts,
            convert_to_tensor=True,
            device=self.device,
            show_progress_bar=False
        )

        sims      = util.cos_sim(emb_gen, emb_truth)
        sem_loss  = torch.mean(1.0 - sims.diag())

        loss = ce_loss + self.sem_coef * sem_loss
        return loss, outputs.logits


# ----------------------------
# 3. Training & Eval
# ----------------------------
def train_epoch(model, loader, optim, scheduler=None, epoch=0):
    model.train()
    tot_loss, tot_acc, n = 0.0, 0.0, 0
    loop = tqdm(loader, desc=f"Epoch {epoch} [Train]", leave=False)
    for batch in loop:
        full    = batch['full_img'].to(model.device)
        patches = batch['patches'].to(model.device)
        ids     = batch['input_ids'].to(model.device)
        mask    = batch['attention_mask'].to(model.device)
        clean   = batch['clean_text']

        optim.zero_grad()
        loss, logits = model(full, patches, ids, mask, clean, labels=ids)
        loss.backward()
        optim.step()
        if scheduler: scheduler.step()

        preds = logits.argmax(-1)
        acc = (preds == ids).masked_select(mask.bool()).float().mean().item()

        bsz = ids.size(0)
        tot_loss += loss.item() * bsz
        tot_acc  += acc * bsz
        n       += bsz
        loop.set_postfix(loss=loss.item(), acc=acc)

    return tot_loss / n, tot_acc / n

@torch.no_grad()
def eval_epoch(model, loader, epoch=0):
    model.eval()
    tot_loss, tot_acc, n = 0.0, 0.0, 0
    loop = tqdm(loader, desc=f"Epoch {epoch} [Val]  ", leave=False)
    for batch in loop:
        full    = batch['full_img'].to(model.device)
        patches = batch['patches'].to(model.device)
        ids     = batch['input_ids'].to(model.device)
        mask    = batch['attention_mask'].to(model.device)
        clean   = batch['clean_text']

        loss, logits = model(full, patches, ids, mask, clean, labels=ids)
        preds = logits.argmax(-1)
        acc = (preds == ids).masked_select(mask.bool()).float().mean().item()

        bsz = ids.size(0)
        tot_loss += loss.item() * bsz
        tot_acc  += acc * bsz
        n       += bsz
        loop.set_postfix(loss=loss.item(), acc=acc)

    return tot_loss / n, tot_acc / n


# ----------------------------
# 4. Plotting
# ----------------------------
def plot_metrics(train_losses, val_losses, train_accs, val_accs):
    epochs = range(1, len(train_losses) + 1)
    plt.figure()
    plt.plot(epochs, train_losses)
    plt.plot(epochs, val_losses)
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train','Val'])
    plt.show()

    plt.figure()
    plt.plot(epochs, train_accs)
    plt.plot(epochs, val_accs)
    plt.title('Token-Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train','Val'])
    plt.show()


# ----------------------------
# 5. Main
# ----------------------------
def main():
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s | %(levelname)s | %(message)s',
        handlers=[
            logging.FileHandler("training.log"),
            logging.StreamHandler()
        ]
    )

    JSON_PATH = './final_samples_both_only_v2_cleaned.json'
    DEVICE    = 'cuda' if torch.cuda.is_available() else 'cpu'
    BATCH, EPOCHS, LR = 8, 5, 1e-4

    # tokenizer
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token': '<|pad|>'})
    tokenizer.bos_token = tokenizer.eos_token

    # data & splits: 95% train, 5% temp âž” split temp 50/50 into val/test (2.5% each)
    ds = FinalSamplesDataset(JSON_PATH, tokenizer)
    idxs = list(range(len(ds)))
    tr, tmp = train_test_split(idxs, test_size=0.05, random_state=42)
    va, te  = train_test_split(tmp, test_size=0.5, random_state=42)

    train_loader = DataLoader(torch.utils.data.Subset(ds, tr),
                              batch_size=BATCH, shuffle=True, collate_fn=collate_fn)
    val_loader   = DataLoader(torch.utils.data.Subset(ds, va),
                              batch_size=BATCH, shuffle=False, collate_fn=collate_fn)
    test_loader  = DataLoader(torch.utils.data.Subset(ds, te),
                              batch_size=1, shuffle=False, collate_fn=collate_fn)

    # model & optimizer
    model     = MedicalReportGenerator(tokenizer, device=DEVICE)
    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)

    # train & validate
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    for ep in range(1, EPOCHS + 1):
        tr_l, tr_a = train_epoch(model, train_loader, optimizer, epoch=ep)
        va_l, va_a = eval_epoch(model, val_loader, epoch=ep)
        print(f"Epoch {ep:02d} â†’ train_loss={tr_l:.4f}, val_loss={va_l:.4f}, train_acc={tr_a:.4f}, val_acc={va_a:.4f}")
        train_losses.append(tr_l)
        val_losses.append(va_l)
        train_accs.append(tr_a)
        val_accs.append(va_a)

    plot_metrics(train_losses, val_losses, train_accs, val_accs)

    # ----------------------------
    # 6. Test: sampling + overall similarity
    # ----------------------------
    model.eval()

    print("\n20 random examples:")
    examples = random.sample(range(len(te)), 20)
    for idx in examples:
        batch = next(iter(DataLoader([ds[te[idx]]], batch_size=1, collate_fn=collate_fn)))
        full_img  = batch['full_img'].to(DEVICE)
        patches   = batch['patches'].to(DEVICE)
        clean_txt = batch['clean_text'][0]

        vt = model.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=DEVICE)

        prefix = tokenizer(f"{tokenizer.bos_token} FINDINGS:", return_tensors='pt').to(DEVICE)
        out = model.decoder.generate(
            **prefix,
            max_length=128,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=1.0,
            num_return_sequences=1,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            use_cache=True
        )
        gen = tokenizer.decode(out[0], skip_special_tokens=True)

        emb = model.embedder.encode([gen, clean_txt], convert_to_tensor=True, device=DEVICE)
        sim = util.cos_sim(emb[0], emb[1]).item()

        print("â€”" * 40)
        print("Clean     :", clean_txt)
        print("Generated :", gen)
        print(f"Similarity: {sim:.4f}")

    # overall average
    sims = []
    for batch in tqdm(test_loader, desc="Evaluating test set"):
        full_img  = batch['full_img'].to(DEVICE)
        patches   = batch['patches'].to(DEVICE)
        clean_txt = batch['clean_text'][0]

        vt = model.encoder(full_img, patches)
        vm = torch.ones(vt.size()[:2], device=DEVICE)

        prefix = tokenizer(f"{tokenizer.bos_token} FINDINGS:", return_tensors='pt').to(DEVICE)
        out = model.decoder.generate(
            **prefix,
            max_length=128,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=1.0,
            num_return_sequences=1,
            encoder_hidden_states=vt,
            encoder_attention_mask=vm,
            use_cache=True
        )
        gen = tokenizer.decode(out[0], skip_special_tokens=True)
        emb_pair = model.embedder.encode([gen, clean_txt], convert_to_tensor=True, device=DEVICE)
        sims.append(util.cos_sim(emb_pair[0], emb_pair[1]).item())

    overall = float(np.mean(sims))
    logging.info(f"\nAverage semantic similarity over {len(sims)} test samples: {overall:.4f}")

if __name__ == '__main__':
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p=paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side,p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out,_ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2,Cg,Hg,Wg = g.shape
        assert B2==B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.1):
    enc.train(); dec.train()
    total_loss=0; total_acc=0
    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        labels = input_ids.clone()
        labels[attention_mask==0] = -100

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)
        cos       = util.cos_sim(gen_embs, ref_embs).diag()
        sim_pen   = (1 - cos).mean()

        loss = ce_loss + sim_weight*sim_pen
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss=0; total_acc=0
    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask==0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =5

    # --- tokenizer & sentence model ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-MiniLM-L6-v2')

    # --- dataset & loaders ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)

    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- training ---
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(enc, dec, tokenizer,
                                      train_loader, optim, sched,
                                      sentencemodel, device,
                                      tokenizer.pad_token_id)
        va_loss, va_acc = eval_epoch(enc, dec, tokenizer,
                                     val_loader, sentencemodel,
                                     device, tokenizer.pad_token_id)

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- plot metrics ---
    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs, label='Train'); plt.plot(epochs, val_accs, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    # --- test & generation with sampling ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)
    
    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # --- print a few examples ---
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p=paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side,p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out,_ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2,Cg,Hg,Wg = g.shape
        assert B2==B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.1):
    enc.train(); dec.train()
    total_loss=0; total_acc=0
    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        labels = input_ids.clone()
        labels[attention_mask==0] = -100

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)
        cos       = util.cos_sim(gen_embs, ref_embs).diag()
        sim_pen   = (1 - cos).mean()

        loss = ce_loss + sim_weight*sim_pen
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss=0; total_acc=0
    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask==0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =1

    # --- tokenizer & higher-capacity sentence model ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    # swap in a stronger embedder:
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders (unchanged) ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler (unchanged) ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)
    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- training with sim_weight=0.2 ---
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.2   # up from 0.1
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- plot metrics ---
    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs, label='Train'); plt.plot(epochs, val_accs, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    # --- test & generation with sampling ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)
    
    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # --- print a few examples ---
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p=paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side,p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out,_ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2,Cg,Hg,Wg = g.shape
        assert B2==B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.1):
    enc.train(); dec.train()
    total_loss=0; total_acc=0
    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        labels = input_ids.clone()
        labels[attention_mask==0] = -100

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)
        cos       = util.cos_sim(gen_embs, ref_embs).diag()
        sim_pen   = (1 - cos).mean()

        loss = ce_loss + sim_weight*sim_pen
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss=0; total_acc=0
    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask==0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =5

    # --- tokenizer & higher-capacity sentence model ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    # swap in a stronger embedder:
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders (unchanged) ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler (unchanged) ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)
    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- training with sim_weight=0.2 ---
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.2   # up from 0.1
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- plot metrics ---
    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs, label='Train'); plt.plot(epochs, val_accs, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    # --- test & generation with sampling ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)
    
    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # --- print a few examples ---
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p=paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side,p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out,_ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2,Cg,Hg,Wg = g.shape
        assert B2==B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.1):
    enc.train(); dec.train()
    total_loss=0; total_acc=0
    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        labels = input_ids.clone()
        labels[attention_mask==0] = -100

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)
        cos       = util.cos_sim(gen_embs, ref_embs).diag()
        sim_pen   = (1 - cos).mean()

        loss = ce_loss + sim_weight*sim_pen
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss=0; total_acc=0
    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask==0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =1

    # --- tokenizer & higher-capacity sentence model ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    # swap in a stronger embedder:
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders (unchanged) ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler (unchanged) ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)
    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- training with sim_weight=0.2 ---
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.2   # up from 0.1
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- plot metrics ---
    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs, label='Train'); plt.plot(epochs, val_accs, label='Val'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    # --- test & generation with sampling ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)
    
    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # strip out the FINDINGS: token
    clean_gen = [g.replace("FINDINGS:", "", 1).strip() for g in all_gen]

    # save full testâ€set results to CSV
    import pandas as pd
    results_df = pd.DataFrame({
        'Reference': all_clean,
        'Generated': clean_gen,
        'Similarity': all_sim
    })
    results_df.to_csv('test_results.csv', index=False)
    print("â†’ Saved full testâ€set results (without 'FINDINGS:') to test_results.csv\n")


    # --- print a few examples ---
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.7, sim_temp=0.1):
    """
    sim_weight: weight on the contrastive InfoNCE loss
    sim_temp: temperature scaling for contrastive loss
    """
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        # prepare labels for CE
        labels = input_ids.clone()
        labels[attention_mask == 0] = -100

        optimizer.zero_grad()
        # forward through encoder + decoder
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        # compute embeddings for contrastive loss
        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        # InfoNCE contrastive loss
        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp   # [B, B]
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        # combined loss
        loss = ce_loss + sim_weight * contrastive_loss

        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask == 0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        = 5e-5
    EPOCHS    = 3

    # --- tokenizer & sentence embedder ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8 * n)
    val_n   = int(0.1 * n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n, val_n, test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)
    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()),
        lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )

    # --- training with stronger contrastive alignment ---
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS + 1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.7,   # stronger alignment weight
            sim_temp=0.1      # contrastive temperature
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- plot metrics ---
    epochs = list(range(1, EPOCHS+1))
    plt.figure()
    plt.plot(epochs, train_losses, label='Train')
    plt.plot(epochs, val_losses,   label='Val')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(epochs, train_accs, label='Train')
    plt.plot(epochs, val_accs,   label='Val')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    # --- test & generation with sampling ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # strip out the FINDINGS: token
    clean_gen = [g.replace("FINDINGS:", "", 1).strip() for g in all_gen]

    # save full testâ€set results to CSV
    import pandas as pd
    results_df = pd.DataFrame({
        'Reference': all_clean,
        'Generated': clean_gen,
        'Similarity': all_sim
    })
    #results_df.to_csv('test_results.csv', index=False)
    print("â†’ Saved full testâ€set results (without 'FINDINGS:') to test_results.csv\n")

    # --- print a few examples ---
    for i in random.sample(range(len(all_clean)), min(20, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-" * 40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.7, sim_temp=0.1):
    """
    sim_weight: weight on the contrastive InfoNCE loss
    sim_temp: temperature scaling for contrastive loss
    """
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        # prepare labels for CE
        labels = input_ids.clone()
        labels[attention_mask == 0] = -100

        optimizer.zero_grad()
        # forward through encoder + decoder
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        # compute embeddings for contrastive loss
        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        # InfoNCE contrastive loss
        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp   # [B, B]
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        # combined loss
        loss = ce_loss + sim_weight * contrastive_loss

        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask == 0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        = 5e-5
    EPOCHS    = 5

    # --- tokenizer & sentence embedder ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8 * n)
    val_n   = int(0.1 * n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n, val_n, test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)
    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()),
        lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )

    # --- training with stronger contrastive alignment ---
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS + 1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.7,   # stronger alignment weight
            sim_temp=0.1      # contrastive temperature
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- plot metrics ---
    epochs = list(range(1, EPOCHS+1))
    plt.figure()
    plt.plot(epochs, train_losses, label='Train')
    plt.plot(epochs, val_losses,   label='Val')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(epochs, train_accs, label='Train')
    plt.plot(epochs, val_accs,   label='Val')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    # --- test & generation with sampling ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # strip out the FINDINGS: token
    clean_gen = [g.replace("FINDINGS:", "", 1).strip() for g in all_gen]

    # save full testâ€set results to CSV
    import pandas as pd
    results_df = pd.DataFrame({
        'Reference': all_clean,
        'Generated': clean_gen,
        'Similarity': all_sim
    })
    results_df.to_csv('test_results.csv', index=False)
    print("â†’ Saved full testâ€set results (without 'FINDINGS:') to test_results.csv\n")

    # --- print a few examples ---
    for i in random.sample(range(len(all_clean)), min(20, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-" * 40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt


# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        # --- Prompt-engineering fix starts here ---
        # A) Tokenize only the fixed prefix as input
        prefix = f"{self.tokenizer.bos_token} FINDINGS:"
        prefix_tok = self.tokenizer(prefix, return_tensors='pt')

        # B) Tokenize the cleaned report + EOS as target
        target = f" {e['clean']} {self.tokenizer.eos_token}"
        target_tok = self.tokenizer(target,
                                     truncation=True,
                                     max_length=self.max_length,
                                     return_tensors='pt')

        # C) Concatenate prefix + target to form the full sequence
        input_ids = torch.cat([prefix_tok.input_ids, target_tok.input_ids], dim=1).squeeze(0)
        attention_mask = torch.cat([prefix_tok.attention_mask, target_tok.attention_mask], dim=1).squeeze(0)

        # D) Build labels that ignore the prefix positions
        labels = input_ids.clone()
        labels[: prefix_tok.input_ids.size(1)] = -100   # mask out BOS and "FINDINGS:"
        # the remaining tokens (the cleaned report + EOS) are used for CE loss

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'labels':         labels,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. COLLATE FN
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids       = [b['input_ids']      for b in batch]
    attention_masks = [b['attention_mask'] for b in batch]
    labels          = [b['labels']         for b in batch]

    input_ids       = nn.utils.rnn.pad_sequence(input_ids,      batch_first=True, padding_value=pad_token_id)
    attention_masks = nn.utils.rnn.pad_sequence(attention_masks,batch_first=True, padding_value=0)
    labels          = nn.utils.rnn.pad_sequence(labels,         batch_first=True, padding_value=-100)

    imgs    = torch.stack([b['full_img']   for b in batch], 0)
    patches = torch.stack([b['patches']    for b in batch], 0)
    cleans  = [b['clean_text']             for b in batch]

    return {
        'input_ids':       input_ids,
        'attention_mask':  attention_masks,
        'labels':          labels,
        'full_img':        imgs,
        'patches':         patches,
        'clean_texts':     cleans
    }


# ----------------------------
# 5. TRAIN / EVAL LOOPS
# ----------------------------
def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.7, sim_temp=0.1):
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['labels'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        # contrastive loss
        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        loss = ce_loss + sim_weight * contrastive_loss
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss / len(loader), total_acc / len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels         = batch['labels'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss / len(loader), total_acc / len(loader)


# ----------------------------
# 6. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        = 5e-5
    EPOCHS    = 1

    # tokenizer & sentence embedder
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # dataset & loaders
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8 * n)
    val_n   = int(0.1 * n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n, val_n, test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # models + optimizer/scheduler
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)
    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()),
        lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )

    # training loop
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS + 1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.7,
            sim_temp=0.1
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # plot metrics
    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs,  label='Train'); plt.plot(epochs, val_accs,  label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    # test & generation
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        # prepare the fixed prefix token IDs once
        prefix_ids = tokenizer(" FINDINGS:", return_tensors="pt").input_ids.to(device)  # shape (1, L)
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = prefix_ids.expand(B, -1)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    # save results
    import pandas as pd
    clean_gen = [g.replace("FINDINGS:", "", 1).strip() for g in all_gen]
    results_df = pd.DataFrame({
        'Reference': all_clean,
        'Generated': clean_gen,
        'Similarity': all_sim
    })
    #results_df.to_csv('test_results.csv', index=False)
    print("â†’ Saved full testâ€set results to test_results.csv")

    # sample prints
    for i in random.sample(range(len(all_clean)), min(20, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-" * 40)

if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.7, sim_temp=0.1):
    """
    sim_weight: weight on the contrastive InfoNCE loss
    sim_temp: temperature scaling for contrastive loss
    """
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        # prepare labels for CE
        labels = input_ids.clone()
        labels[attention_mask == 0] = -100

        optimizer.zero_grad()
        # forward through encoder + decoder
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        # compute embeddings for contrastive loss
        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        # InfoNCE contrastive loss
        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp   # [B, B]
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        # combined loss
        loss = ce_loss + sim_weight * contrastive_loss

        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask == 0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    # --- setup ---
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =1

    # --- tokenizer & sentence embedder ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(
        tokenizer,
        decoder_name='gpt2',
        fusion_dim=768
    ).to(device)

    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- training loop ---
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.2
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )
        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- deterministic decoding on test set ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []

    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Deterministic Decode"):
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            # start with BOS token
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id,
                                     dtype=torch.long, device=device)

            # deterministic beam search
            outputs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=False,           # turn off sampling
                num_beams=5,               # number of beams; set to 1 for greedy
                early_stopping=True,       # stop when all beams hit EOS
                no_repeat_ngram_size=2     # optional repetition penalty
            )

            gens = tokenizer.batch_decode(outputs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity (deterministic): {avg_sim:.4f}\n")

    # print a few examples
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        # resize embeddings for new pad token
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.7, sim_temp=0.1):
    """
    sim_weight: weight on the contrastive InfoNCE loss
    sim_temp: temperature scaling for contrastive loss
    """
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        # prepare labels for CE
        labels = input_ids.clone()
        labels[attention_mask == 0] = -100

        optimizer.zero_grad()
        # forward through encoder + decoder
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        # compute embeddings for contrastive loss
        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        # InfoNCE contrastive loss
        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp   # [B, B]
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        # combined loss
        loss = ce_loss + sim_weight * contrastive_loss

        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask == 0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    avg_loss = total_loss / len(loader)
    avg_acc  = total_acc  / len(loader)
    return avg_loss, avg_acc


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    # --- setup ---
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =5   # increase total epochs

    # --- tokenizer & sentence embedder ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(
        tokenizer,
        decoder_name='gpt2',
        fusion_dim=768
    ).to(device)

    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- 1) Warm up cross-attention layers only ---
    # Freeze all pretrained GPT2 parameters except cross-attn
    for name, p in dec.decoder.named_parameters():
        if "crossattention" not in name:
            p.requires_grad = False

    print("=== Warmup: training cross-attention only ===")
    for epoch in range(1, 4):  # e.g. 3 warmup epochs
        tr_loss, _ = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.0   # no similarity penalty yet
        )
        va_loss, _ = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )
        print(f"[Warmup Epoch {epoch}/3] Train loss {tr_loss:.4f} | Val loss {va_loss:.4f}")

    # --- 2) Unfreeze all and lower LR ---
    for p in dec.decoder.parameters():
        p.requires_grad = True
    for g in optim.param_groups:
        g["lr"] = 1e-5
    print("=== Full fine-tuning: unfreeze all parameters ===")

    # --- 3) Full train & validate ---
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.2
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )
        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- 4) Deterministic decoding on test set ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []

    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Deterministic Decode"):
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id,
                                     dtype=torch.long, device=device)

            outputs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=False,
                num_beams=5,
                early_stopping=True,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outputs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity (deterministic): {avg_sim:.4f}\n")

    # Print sample outputs
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)

if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt

# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        prompt = f"{self.tokenizer.bos_token} FINDINGS: {e['clean']} {self.tokenizer.eos_token}"
        tok = self.tokenizer(
            prompt,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )
        input_ids = tok['input_ids'].squeeze(0)
        attention_mask = tok['attention_mask'].squeeze(0)

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side,p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out,_ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2,Cg,Hg,Wg = g.shape
        assert B2==B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. GPT2 CROSS-ATTENTION DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True

        # Load GPT-2 with cross-attn
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)

        # Resize embeddings to include [PAD]
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        # Project if dims differ
        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask, encodings, labels=None):
        if self.enc_proj is not None:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. HELPERS & LOOPS
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids = [b['input_ids'] for b in batch]
    masks     = [b['attention_mask'] for b in batch]
    input_ids = nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=pad_token_id)
    masks     = nn.utils.rnn.pad_sequence(masks,     batch_first=True, padding_value=0)
    imgs    = torch.stack([b['full_img'] for b in batch], 0)
    patches = torch.stack([b['patches'] for b in batch], 0)
    cleans  = [b['clean_text'] for b in batch]
    return {
        'input_ids':      input_ids,
        'attention_mask': masks,
        'full_img':       imgs,
        'patches':        patches,
        'clean_texts':    cleans
    }

def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.1):
    enc.train(); dec.train()
    total_loss=0; total_acc=0
    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        labels = input_ids.clone()
        labels[attention_mask==0] = -100

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)
        cos       = util.cos_sim(gen_embs, ref_embs).diag()
        sim_pen   = (1 - cos).mean()

        loss = ce_loss + sim_weight*sim_pen
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss=0; total_acc=0
    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            labels = input_ids.clone()
            labels[attention_mask==0] = -100

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss/len(loader), total_acc/len(loader)


# ----------------------------
# 5. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        =5e-5
    EPOCHS    =5

    # --- tokenizer & sentence embedder ---
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # --- dataset & loaders ---
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8*n)
    val_n   = int(0.1*n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n,val_n,test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # --- models & optimizer/scheduler ---
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(
        tokenizer,
        decoder_name='gpt2',
        fusion_dim=768
    ).to(device)

    optim = torch.optim.AdamW(
        list(enc.parameters()) + list(dec.parameters()), lr=LR
    )
    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1*total_steps),
        num_training_steps=total_steps
    )

    # --- train & validate ---
    for epoch in range(1, EPOCHS+1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.2
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )
        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # --- test & generation ---
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = torch.full((B,1), tokenizer.bos_token_id, dtype=torch.long, device=device)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")
    for i in random.sample(range(len(all_clean)), min(10, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-"*40)


if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt


# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        # --- Prompt-engineering fix starts here ---
        # A) Tokenize only the fixed prefix as input
        prefix = f"{self.tokenizer.bos_token} FINDINGS:"
        prefix_tok = self.tokenizer(prefix, return_tensors='pt')

        # B) Tokenize the cleaned report + EOS as target
        target = f" {e['clean']} {self.tokenizer.eos_token}"
        target_tok = self.tokenizer(target,
                                     truncation=True,
                                     max_length=self.max_length,
                                     return_tensors='pt')

        # C) Concatenate prefix + target to form the full sequence
        input_ids = torch.cat([prefix_tok.input_ids, target_tok.input_ids], dim=1).squeeze(0)
        attention_mask = torch.cat([prefix_tok.attention_mask, target_tok.attention_mask], dim=1).squeeze(0)

        # D) Build labels that ignore the prefix positions
        labels = input_ids.clone()
        labels[: prefix_tok.input_ids.size(1)] = -100   # mask out BOS and "FINDINGS:"

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'labels':         labels,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. COLLATE FN
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids       = [b['input_ids']      for b in batch]
    attention_masks = [b['attention_mask'] for b in batch]
    labels          = [b['labels']         for b in batch]

    input_ids       = nn.utils.rnn.pad_sequence(input_ids,      batch_first=True, padding_value=pad_token_id)
    attention_masks = nn.utils.rnn.pad_sequence(attention_masks,batch_first=True, padding_value=0)
    labels          = nn.utils.rnn.pad_sequence(labels,         batch_first=True, padding_value=-100)

    imgs    = torch.stack([b['full_img']   for b in batch], 0)
    patches = torch.stack([b['patches']    for b in batch], 0)
    cleans  = [b['clean_text']             for b in batch]

    return {
        'input_ids':       input_ids,
        'attention_mask':  attention_masks,
        'labels':          labels,
        'full_img':        imgs,
        'patches':         patches,
        'clean_texts':     cleans
    }


# ----------------------------
# 5. TRAIN / EVAL LOOPS
# ----------------------------
def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.7, sim_temp=0.2):
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['labels'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        loss = ce_loss + sim_weight * contrastive_loss
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss / len(loader), total_acc / len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels         = batch['labels'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss / len(loader), total_acc / len(loader)


# ----------------------------
# 6. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        = 5e-5
    EPOCHS    = 10

    # tokenizer & sentence embedder
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    # dataset & loaders
    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8 * n)
    val_n   = int(0.1 * n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n, val_n, test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    # models + optimizer/scheduler
    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)

    # --- freeze all, then unfreeze only cross-attention layers ---
    for p in enc.parameters(): p.requires_grad = False
    for p in dec.parameters(): p.requires_grad = False

    # encoder fusion cross-attn
    for module in [enc.cross1, enc.cross2]:
        for p in module.parameters(): p.requires_grad = True

    # decoder cross-attn
    for name, p in dec.decoder.named_parameters():
        if "crossattention" in name:
            p.requires_grad = True

    # (optional) unfreeze the encoderâ†’decoder projection
    if dec.enc_proj is not None:
        for p in dec.enc_proj.parameters():
            p.requires_grad = True

    # rebuild optimizer on just the trainable params
    trainable = [p for p in list(enc.parameters()) + list(dec.parameters()) if p.requires_grad]
    optim = torch.optim.AdamW(trainable, lr=1e-5)

    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )

    # training loop
    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS + 1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.7,
            sim_temp=0.1
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    # plot metrics
    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs,  label='Train'); plt.plot(epochs, val_accs,  label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    # test & generation
    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        prefix_ids = tokenizer(" FINDINGS:", return_tensors="pt").input_ids.to(device)
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = prefix_ids.expand(B, -1)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=50,
                top_p=0.9,
                temperature=0.7,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    import pandas as pd
    clean_gen = [g.replace("FINDINGS:", "", 1).strip() for g in all_gen]
    results_df = pd.DataFrame({
        'Reference': all_clean,
        'Generated': clean_gen,
        'Similarity': all_sim
    })
    results_df.to_csv('test_results.csv', index=False)
    print("â†’ Saved full testâ€set results to test_results.csv")

    for i in random.sample(range(len(all_clean)), min(20, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-" * 40)

if __name__ == "__main__":
    main()
```

```python
import os
import json
import random
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models

from transformers import (
    GPT2Tokenizer,
    GPT2Config,
    GPT2LMHeadModel,
    get_linear_schedule_with_warmup
)
from sentence_transformers import SentenceTransformer, util
import matplotlib.pyplot as plt


# ----------------------------
# 1. DATASET
# ----------------------------
class FinalSamplesDataset(Dataset):
    def __init__(self, json_path, tokenizer,
                 image_transform=None, patch_transform=None,
                 max_length=512):
        self.tokenizer = tokenizer
        self.max_length = max_length

        self.image_transform = image_transform or transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])
        self.patch_transform = patch_transform or transforms.Compose([
            transforms.Resize((112, 112)),
            transforms.RandomAffine(degrees=10, translate=(0.1,0.1), scale=(0.95,1.05)),
            transforms.ToTensor()
        ])

        with open(json_path, 'r') as f:
            raw = json.load(f)

        filtered = []
        for item in raw:
            merged = item.get('merged_image_path', '')
            fps = item.get('file_paths', [])
            if isinstance(fps, str):
                fps = [fps]
            paths = [merged] + fps
            if any(os.path.exists(p) for p in paths):
                filtered.append((merged, fps, item))

        self.data = {}
        for i, (merged, fps, item) in enumerate(filtered):
            self.data[i] = {
                'full': merged,
                'patches': fps,
                'keypoints': item.get('keypoints', {}),
                'clean': item.get('cleaned_report', '')
            }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        e = self.data[idx]
        full_path = e['full'] if os.path.exists(e['full']) else e['patches'][0]
        img = Image.open(full_path).convert('RGB')
        img = self.image_transform(img)

        patches = self._gen_patches(e['patches'], e['keypoints'])
        pt = [self.patch_transform(Image.fromarray(p)) for p in patches]
        patches_tensor = torch.stack(pt,0) if pt else torch.zeros(34,3,112,112)

        # Prompt-engineering: prefix only as context
        prefix = f"{self.tokenizer.bos_token} FINDINGS:"
        prefix_tok = self.tokenizer(prefix, return_tensors='pt')

        target = f" {e['clean']} {self.tokenizer.eos_token}"
        target_tok = self.tokenizer(
            target,
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )

        input_ids = torch.cat([prefix_tok.input_ids, target_tok.input_ids], dim=1).squeeze(0)
        attention_mask = torch.cat([prefix_tok.attention_mask, target_tok.attention_mask], dim=1).squeeze(0)

        labels = input_ids.clone()
        labels[: prefix_tok.input_ids.size(1)] = -100

        return {
            'full_img':       img,
            'patches':        patches_tensor,
            'input_ids':      input_ids,
            'attention_mask': attention_mask,
            'labels':         labels,
            'clean_text':     e['clean']
        }

    def _gen_patches(self, paths, kps_dict,
                     crop_size=(200,300), patch_size=(112,112)):
        def extract(arr, side_kps):
            lst = []
            pts = side_kps[0]['keypoints']
            for i in range(17):
                x,y,s = int(pts[3*i]), int(pts[3*i+1]), pts[3*i+2]
                if s>0:
                    x0 = max(x-crop_size[0]//2, 0)
                    y0 = max(y-crop_size[1]//2, 0)
                    x1 = min(x+crop_size[0]//2, arr.shape[1])
                    y1 = min(y+crop_size[1]//2, arr.shape[0])
                    c = arr[y0:y1, x0:x1]
                    if c.size:
                        lst.append(cv2.resize(c, patch_size))
            return lst

        def pad17(lst):
            black = np.zeros((patch_size[1], patch_size[0], 3), np.uint8)
            while len(lst) < 17:
                lst.append(black)
            return lst[:17]

        left, right = [], []
        if len(paths)==1:
            p = paths[0]
            if os.path.exists(p):
                arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                if kps_dict.get('left'):
                    left  = extract(arr, kps_dict['left'])
                if kps_dict.get('right'):
                    right = extract(arr, kps_dict['right'])
        else:
            for side, p in zip(['left','right'], paths):
                if p and os.path.exists(p):
                    arr = cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB)
                    if kps_dict.get(side):
                        lst = extract(arr, kps_dict[side])
                        (left if side=='left' else right).extend(lst)

        if left and not right:
            right = [cv2.flip(p,1) for p in left]
        if right and not left:
            left  = [cv2.flip(p,1) for p in right]
        if not left and not right:
            return pad17([])+pad17([])
        return pad17(left)+pad17(right)


# ----------------------------
# 2. FUSION ENCODER
# ----------------------------
class CrossAttentionBlock(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.mha  = nn.MultiheadAttention(embed_dim=dim, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim)
        self.ff   = nn.Sequential(
            nn.Linear(dim, dim*4),
            nn.GELU(),
            nn.Linear(dim*4, dim),
        )

    def forward(self, q, k, v, mask=None):
        attn_out, _ = self.mha(query=q, key=k, value=v, key_padding_mask=mask)
        x = self.norm(q + attn_out)
        x = self.norm(x + self.ff(x))
        return x

class VisionFusionEncoder(nn.Module):
    def __init__(self, swin_model_name='swin_base_patch4_window7_224',
                 fusion_dim=768):
        super().__init__()
        import timm
        self.global_encoder = timm.create_model(
            swin_model_name, pretrained=True, features_only=True
        )
        num_ch = self.global_encoder.feature_info[-1]['num_chs']
        self.global_proj = nn.Linear(num_ch, fusion_dim)

        resnet = models.resnet50(pretrained=True)
        self.local_encoder = nn.Sequential(*list(resnet.children())[:-1])
        self.local_proj    = nn.Linear(2048, fusion_dim)

        self.cross1 = CrossAttentionBlock(fusion_dim)
        self.cross2 = CrossAttentionBlock(fusion_dim)

    def forward(self, full_img, patches):
        B, Np, C, pH, pW = patches.shape

        feats = self.global_encoder(full_img)
        g = feats[-1]
        if g.ndim==4 and g.shape[1]<g.shape[-1]:
            g = g.permute(0,3,1,2)
        B2, Cg, Hg, Wg = g.shape
        assert B2 == B
        g = g.flatten(2).permute(0,2,1)
        g = self.global_proj(g)

        p = patches.view(-1, C, pH, pW)
        lp = self.local_encoder(p).squeeze(-1).squeeze(-1)
        lp = self.local_proj(lp)
        l  = lp.view(B, Np, -1)

        g2 = self.cross1(g, l, l)
        l2 = self.cross2(l, g, g)

        fusion = torch.cat([g2, l2], dim=1)
        return fusion


# ----------------------------
# 3. DECODER
# ----------------------------
class GPT2CrossAttentionDecoder(nn.Module):
    def __init__(self, tokenizer, decoder_name='gpt2', fusion_dim=768):
        super().__init__()
        cfg = GPT2Config.from_pretrained(decoder_name)
        cfg.add_cross_attention = True
        self.decoder = GPT2LMHeadModel.from_pretrained(decoder_name, config=cfg)
        self.decoder.resize_token_embeddings(len(tokenizer))
        self.decoder.config.pad_token_id = tokenizer.pad_token_id

        if fusion_dim != cfg.hidden_size:
            self.enc_proj = nn.Linear(fusion_dim, cfg.hidden_size)
        else:
            self.enc_proj = None

    def forward(self, input_ids, attention_mask,
                encodings, labels=None):
        if self.enc_proj:
            encodings = self.enc_proj(encodings)
        return self.decoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encoder_hidden_states=encodings,
            encoder_attention_mask=None,
            labels=labels,
            return_dict=True
        )


# ----------------------------
# 4. COLLATE FN
# ----------------------------
def collate_fn(batch, pad_token_id):
    input_ids       = [b['input_ids']      for b in batch]
    attention_masks = [b['attention_mask'] for b in batch]
    labels          = [b['labels']         for b in batch]

    input_ids       = nn.utils.rnn.pad_sequence(input_ids,      batch_first=True, padding_value=pad_token_id)
    attention_masks = nn.utils.rnn.pad_sequence(attention_masks,batch_first=True, padding_value=0)
    labels          = nn.utils.rnn.pad_sequence(labels,         batch_first=True, padding_value=-100)

    imgs    = torch.stack([b['full_img']   for b in batch], 0)
    patches = torch.stack([b['patches']    for b in batch], 0)
    cleans  = [b['clean_text']             for b in batch]

    return {
        'input_ids':       input_ids,
        'attention_mask':  attention_masks,
        'labels':          labels,
        'full_img':        imgs,
        'patches':         patches,
        'clean_texts':     cleans
    }


# ----------------------------
# 5. TRAIN / EVAL LOOPS
# ----------------------------
def compute_token_acc(logits, labels, ignore_index=-100):
    preds = logits.argmax(-1)
    mask  = labels != ignore_index
    correct = (preds == labels) & mask
    return correct.sum().item() / mask.sum().item()

def train_epoch(enc, dec, tokenizer, loader, optimizer, scheduler,
                sentencemodel, device, pad_id, sim_weight=0.3, sim_temp=0.1):
    enc.train(); dec.train()
    total_loss = 0.0
    total_acc  = 0.0

    for batch in tqdm(loader, desc="Train"):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['labels'].to(device)
        imgs           = batch['full_img'].to(device)
        patches        = batch['patches'].to(device)
        clean_texts    = batch['clean_texts']

        optimizer.zero_grad()
        encodings = enc(imgs, patches)
        outputs   = dec(
            input_ids=input_ids,
            attention_mask=attention_mask,
            encodings=encodings,
            labels=labels
        )
        ce_loss = outputs.loss
        logits  = outputs.logits
        acc     = compute_token_acc(logits, labels)

        with torch.no_grad():
            ref_embs = sentencemodel.encode(clean_texts, convert_to_tensor=True).to(device)
        gen_ids   = logits.argmax(-1)
        gen_texts = tokenizer.batch_decode(gen_ids.cpu(), skip_special_tokens=True)
        gen_embs  = sentencemodel.encode(gen_texts, convert_to_tensor=True).to(device)

        sims = util.cos_sim(gen_embs, ref_embs) / sim_temp
        labels_contrast = torch.arange(sims.size(0), device=device)
        contrastive_loss = nn.CrossEntropyLoss()(sims, labels_contrast)

        loss = ce_loss + sim_weight * contrastive_loss
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        total_acc  += acc

    return total_loss / len(loader), total_acc / len(loader)

def eval_epoch(enc, dec, tokenizer, loader, sentencemodel, device, pad_id):
    enc.eval(); dec.eval()
    total_loss = 0.0
    total_acc  = 0.0

    with torch.no_grad():
        for batch in tqdm(loader, desc="Eval"):
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels         = batch['labels'].to(device)
            imgs           = batch['full_img'].to(device)
            patches        = batch['patches'].to(device)

            encodings = enc(imgs, patches)
            outputs   = dec(
                input_ids=input_ids,
                attention_mask=attention_mask,
                encodings=encodings,
                labels=labels
            )
            ce_loss = outputs.loss
            logits  = outputs.logits
            acc     = compute_token_acc(logits, labels)

            total_loss += ce_loss.item()
            total_acc  += acc

    return total_loss / len(loader), total_acc / len(loader)


# ----------------------------
# 6. MAIN
# ----------------------------
def main():
    device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    DATA_JSON = 'final_samples_both_only_v2_cleaned.json'
    BATCH_SIZE=4
    LR        = 5e-5
    EPOCHS    = 10

    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    tokenizer.add_special_tokens({'pad_token':'[PAD]'})
    tokenizer.padding_side = 'left'
    sentencemodel = SentenceTransformer('all-mpnet-base-v2')

    ds = FinalSamplesDataset(DATA_JSON, tokenizer)
    n  = len(ds)
    train_n = int(0.8 * n)
    val_n   = int(0.1 * n)
    test_n  = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(ds, [train_n, val_n, test_n])

    coll = lambda b: collate_fn(b, tokenizer.pad_token_id)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  collate_fn=coll)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, collate_fn=coll)

    enc = VisionFusionEncoder().to(device)
    dec = GPT2CrossAttentionDecoder(tokenizer).to(device)

    # 1) Freeze everything
    for p in enc.parameters(): p.requires_grad = False
    for p in dec.parameters(): p.requires_grad = False

    # 2) Unfreeze encoder fusion cross-attn
    for m in [enc.cross1, enc.cross2]:
        for p in m.parameters(): p.requires_grad = True

    # 3) Unfreeze decoder cross-attn
    for name, p in dec.decoder.named_parameters():
        if "crossattention" in name:
            p.requires_grad = True

    # 4) Unfreeze the encoderâ†’decoder projection
    if dec.enc_proj is not None:
        for p in dec.enc_proj.parameters():
            p.requires_grad = True

    # 5) Unfreeze the final LM head
    for p in dec.decoder.lm_head.parameters():
        p.requires_grad = True

    # 6) Unfreeze the last 2 transformer blocks
    total_blocks = len(dec.decoder.transformer.h)
    for i in [total_blocks-2, total_blocks-1]:
        for p in dec.decoder.transformer.h[i].parameters():
            p.requires_grad = True

    # 7) Build optimizer on just the trainable params
    trainable = [p for p in list(enc.parameters()) + list(dec.parameters()) if p.requires_grad]
    optim = torch.optim.AdamW(trainable, lr=5e-6)

    total_steps = EPOCHS * len(train_loader)
    sched = get_linear_schedule_with_warmup(
        optim,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )

    train_losses, val_losses = [], []
    train_accs,   val_accs   = [], []
    for epoch in range(1, EPOCHS + 1):
        tr_loss, tr_acc = train_epoch(
            enc, dec, tokenizer,
            train_loader, optim, sched,
            sentencemodel, device,
            tokenizer.pad_token_id,
            sim_weight=0.3,
            sim_temp=0.1
        )
        va_loss, va_acc = eval_epoch(
            enc, dec, tokenizer,
            val_loader, sentencemodel,
            device, tokenizer.pad_token_id
        )

        train_losses.append(tr_loss)
        val_losses.append(va_loss)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)

        print(f"Epoch {epoch}/{EPOCHS} | "
              f"Train loss {tr_loss:.4f}, acc {tr_acc:.4f} | "
              f"Val   loss {va_loss:.4f}, acc {va_acc:.4f}")

    epochs = list(range(1, EPOCHS+1))
    plt.figure(); plt.plot(epochs, train_losses, label='Train'); plt.plot(epochs, val_losses, label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
    plt.figure(); plt.plot(epochs, train_accs,  label='Train'); plt.plot(epochs, val_accs,  label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

    enc.eval(); dec.eval()
    all_clean, all_gen, all_sim = [], [], []
    with torch.no_grad():
        prefix_ids = tokenizer(" FINDINGS:", return_tensors="pt").input_ids.to(device)
        for batch in test_loader:
            imgs    = batch['full_img'].to(device)
            patches = batch['patches'].to(device)
            cleans  = batch['clean_texts']
            encs    = enc(imgs, patches)

            B = imgs.size(0)
            prompt_ids = prefix_ids.expand(B, -1)

            outs = dec.decoder.generate(
                input_ids=prompt_ids,
                max_length=128,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id,
                encoder_hidden_states=encs,
                do_sample=True,
                top_k=20,
                top_p=0.7,
                temperature=0.5,
                num_beams=3,
                no_repeat_ngram_size=2
            )

            gens = tokenizer.batch_decode(outs, skip_special_tokens=True)
            sims = util.cos_sim(
                sentencemodel.encode(gens, convert_to_tensor=True).to(device),
                sentencemodel.encode(cleans, convert_to_tensor=True).to(device)
            ).diag().cpu().tolist()

            all_clean.extend(cleans)
            all_gen.extend(gens)
            all_sim.extend(sims)

    avg_sim = sum(all_sim) / len(all_sim) if all_sim else 0.0
    print(f"\nOverall average similarity: {avg_sim:.4f}\n")

    import pandas as pd
    clean_gen = [g.replace("FINDINGS:", "", 1).strip() for g in all_gen]
    results_df = pd.DataFrame({
        'Reference': all_clean,
        'Generated': clean_gen,
        'Similarity': all_sim
    })
    results_df.to_csv('test_results.csv', index=False)
    print("â†’ Saved full testâ€set results to test_results.csv")

    for i in random.sample(range(len(all_clean)), min(20, len(all_clean))):
        print("REF:", all_clean[i])
        print("GEN:", all_gen[i])
        print(f"SIM: {all_sim[i]:.4f}")
        print("-" * 40)

if __name__ == "__main__":
    main()
```
